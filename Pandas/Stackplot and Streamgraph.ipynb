```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Year = [1960, 1970, 1980, 1990, 2000, 2010, 2020]
Changes = {
    "Bare Soil" : [0.225, 0.311, 0.4284, 0.534, 0.623, 0.723, 0.8474],
    "Built Up" : [0.325, 0.411, 0.5284, 0.634, 0.723, 0.823, 0.9474],
    "Water" : [1.525, 1.711, 1.9284, 2.734, 2.823, 2.923, 3.2474],
    "Forest" : [0.425, 0.511, 0.6284, 0.734, 0.823, 0.923, 1.2474],
    "Farmland" : [0.525, 0.711, 0.9284, 1.734, 1.823, 2.923, 2.2474],
    
}

fig, ax = plt.subplots()
ax.stackplot(Year, Changes.values(),
             labels= Changes.keys(), alpha =1)
ax.legend(loc= 0,  reverse  =True)
ax.set_title("Land Cover Changes from 1960 to 2020")
ax.set_xlabel("Years at Decade Intervals")
ax.set_ylabel("Area  (Ha)")

plt.show()
```

```python
np.random.seed(19602020)

def Steph(x, n = 5):
    """ Return the random of the *n* Steph, also Evalute"""
    def add_random_Steph(a):
        amp = 1 /  (.1 + np.random.random())
        dx = x[-1] - x[0]
        x0 = (2 * np.random.random() - .5) * dx
        z = 10 / (.1 + np.random.random()) / dx
        a += amp * np.exp(-(z * (x - x0)) **2)
    a = np.zeros_like(x)
    for j in range(n):
         add_random_Steph(a)
    return(a)
x = np.linspace(0, 100, 10)
ys = [Steph(x) for _ in range(3)]


fig, ax = plt.subplots()
ax.stackplot(x, ys, baseline = "weighted_wiggle")
plt.show()
                          
                      
```

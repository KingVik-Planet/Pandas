```python
import matplotlib.pyplot as plt
import numpy as np

Class = ("Built Area", "Water", "Bare soil", "Vegetation", "Urban")
Year = {
    "2000" : (14.98, 11.4, 30.23, 17.4, 25.5),
    "2010" : (26.98, 20.4, 19.23, 21.9, 33.4),
    "2020" : (47.98, 40.4, 40.23, 60.3, 35.1),
}
x = np.arange(len(Class))
width = 0.3
multiplier = 0

fig, ax = plt.subplots(layout = "constrained")

for attribute, measurement in Year.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label= attribute)
    ax.bar_label(rects, padding = 3)
    multiplier += 1
ax.set_ylabel("Area in Ha")
ax.set_xlabel("Classes of LCLU")
ax.set_title("Land Cover Land Use Analysis for 2000, 2010 and 2020")
ax.set_xticks(x + width, Class)
ax.legend(loc = "upper left", ncols =3)
ax.set_ylim(0, 70)
ax.grid(color='k', linestyle='-', linewidth=0.1)
plt.show()
```

# Categorical Variable

```python
import matplotlib.pyplot as plt

Class = {"Pablo" : 10, "Chika" : 2, "Sterling" : 34, "Erus" : 46}
name = list(Class.keys())
Networth = list(Class.values())

fig, axs = plt.subplots(1,3, figsize = (9,3), sharey= True)
axs[0].bar(name, Networth)
axs[1].scatter(name, Networth)
axs[2].plot(name, Networth)
```

# Pie Chart

```python
Class = "Built Area", "Water", "Bare soil", "Vegetation", "Urban"
Year_2020 = [47.98, 40.4, 40.23, 60.3, 35.1]
fix, ax = plt.subplots()
ax.pie(Year_2020, labels = Class)
```

```python
fix, ax = plt.subplots()
ax.pie(Year_2020, labels = Class, autopct = "%1.2f")
```

```python
fix, ax = plt.subplots()
ax.pie(Year_2020, labels = Class, autopct = "%1.2f", pctdistance = 1, labeldistance = .5)
```

```python
fix, ax = plt.subplots()
ax.pie(Year_2020, labels = Class, colors=["red", "blue", "yellow", "black", "green"])
```

```python
fix, ax = plt.subplots()
ax.pie(Year_2020, labels = Class, hatch=["**0", "o0", "o.o", ".^.", "v||."])
```

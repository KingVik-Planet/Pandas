# Matplotlib Creating Pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([180,45 ,45 ,45 ,45]) # x /sum(X) formula
plt.pie(x)
plt.show()
```

# Pie Label

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([90,45 ,45 ,45 ,45]) 
mylabels = ["Steph", "Pablo", "Nulu", "Stella", "Sterling"]
plt.pie(x, labels = mylabels)
plt.show()
```

##  Start Angle

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([90,45 ,45 ,45 ,45]) 
mylabels = ["Steph", "Pablo", "Nulu", "Stella", "Sterling"]

plt.pie(x, labels = mylabels, startangle= 45)
plt.show()
```

## Explode

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([90,45 ,45 ,45 ,45]) 
mylabels = ["Steph", "Pablo", "Nulu", "Stella", "Sterling"]
myexplode= [0, 0.2, 0, 0.2, 0]

plt.pie(x, labels = mylabels, startangle= 0, explode = myexplode)
plt.show()
```

## Shadow

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([90,45 ,45 ,45 ,45]) 
mylabels = ["Steph", "Pablo", "Nulu", "Stella", "Sterling"]
myexplode= [0, 1, 0, 0.2, 1]

plt.pie(x, labels = mylabels, startangle= 0, explode = myexplode, shadow = True)
plt.show()
```

# Color 

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([90,45 ,45 ,45 ,45]) 
mylabels = ["Steph", "Pablo", "Nulu", "Stella", "Sterling"]
myexplode= [0, 0.2, 0, 0.2, 0.3]
mycolor = ["black", "blue", "green", "yellow", "red"]
plt.pie(x, labels = mylabels, startangle= 0, explode = myexplode, shadow = True, colors = mycolor)
plt.show()
```

## Legend

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([90,45 ,45 ,45 ,45]) 
mylabels = ["Steph", "Pablo", "Nulu", "Stella", "Sterling"]
myexplode= [0, 0.2, 0, 0.2, 0.3]
mycolor = ["black", "blue", "green", "yellow", "red"]
plt.pie(x, labels = mylabels, startangle= 0, explode = myexplode, shadow = True, colors = mycolor)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([90,45 ,45 ,45 ,45]) 
mylabels = ["Steph", "Pablo", "Nulu", "Stella", "Sterling"]
myexplode= [0, 0.2, 0, 0.2, 0.3]
mycolor = ["black", "blue", "green", "yellow", "red"]
plt.pie(x, labels = mylabels, startangle= 0, explode = myexplode, shadow = True, colors = mycolor)
plt.legend(title = "Begineers")
plt.show()
```

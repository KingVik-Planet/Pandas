```python
import matplotlib.pyplot as plt
import numpy as np

Geomatics = ("Pablo", "Sterling", "Nulu")
Gender = {
    "Male": np.array([70, 40, 56]),
    "Female": np.array([16, 23, 40])
}
width = 0.4

fig, ax = plt.subplots()
bottom = np.zeros(3)

for Gen, Gender in Gender.items():
    Plot= ax.bar(Geomatics, Gender, width, label = Gen, bottom = bottom)
    ax.bar_label(Plot , label_type = "edge")
plt.legend()
ax.set_ylabel("Begineers Value")
ax.set_xlabel("Begineers Name")
ax.set_title("Number of Genders in Geomatics")
plt.show()
    
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Variables - 5 Classes of Land use and Land Cover
LandUse = ("Forest", "Shrubs", "Urban", "Water", "Grassland")

# Class (2000, 2010, and 2020) with Count
# Sum of counts for each year is 54321
LandUseData = {
    "2000": np.array([20000, 15000, 8000, 6321, 5000]),
    "2010": np.array([18000, 16000, 10000, 6321, 4000]),
    "2020": np.array([15000, 17000, 12000, 5321, 5000])
}

width = 0.7

fig, ax = plt.subplots()
bottom = np.zeros(5)

for year, data in LandUseData.items():
    Plot = ax.bar(LandUse, data, width, label=year, bottom=bottom)
    ax.bar_label(Plot, label_type="center")
    bottom += data

plt.legend()
ax.set_ylabel("Area (hectares)")
ax.set_xlabel("Land Use / Land Cover Class")
ax.set_title("Land Use and Land Cover Change (2000, 2010 and 2020)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(123321)
LandUse = ("Forest", "Shrubs", "Urban", "Water", "Grassland")
y_pos = np.arange(len(LandUse))
changes = 4 + 6 * np.random.rand(len(LandUse))
error = np.random.rand(len(LandUse))
fig, ax = plt.subplots()
hbar = ax.barh(y_pos, changes, xerr=error, align="center", color = "g")
ax.set_yticks(y_pos, labels=LandUse)
ax.invert_yaxis()
ax.set_title("Land Use Changes Over Time")
ax.set_xlabel("Changes in Land Use and Cover", ha = "right")
ax.set_ylabel("Land Use Categories")
ax.bar_label(hbar, labels=[f'{c:.2f}%' for c in changes])
ax.set_xlim(right = 15)
plt.figtext(0.9, 0.01, "Data source: https://data.apps.fao.org/", ha="right", fontsize=6)

plt.show()
```
